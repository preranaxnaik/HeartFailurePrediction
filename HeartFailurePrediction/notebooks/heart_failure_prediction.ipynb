```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.value_counts('DEATH_EVENT')
```

```python
dataset.shape
```

```python
X = dataset.drop(['DEATH_EVENT'],axis=1)
Y = dataset['DEATH_EVENT']
```

```python
scale = StandardScaler()
```

```python
scaled_X= scale.fit_transform(X)
```

```python
X_train , X_test , Y_train ,Y_test = train_test_split(scaled_X,Y,test_size=0.2,random_state=31)
```

```python
from sklearn.ensemble import AdaBoostClassifier
# xgboost
from sklearn.tree import DecisionTreeClassifier

model = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=3),  # slightly deeper tree
    n_estimators=75,                               # more weak learners
    learning_rate=0.7,                              # smaller step size
    random_state=30
)
model.fit(X_train, Y_train)

preds = model.predict(X_test)
print("Boosting (AdaBoost) Accuracy:", accuracy_score(Y_test, preds))
```

```python
input=(55.0,	0,	7861,	0	,38	,0	,263358.03,	1.1	,136,	1	,0	,6)
input_as_numpy_array=np.asarray(input)
input_reshaped=input_as_numpy_array.reshape(1,-1)
input_data=scale.transform(input_reshaped)

prediction=model.predict(input_data)
print(prediction)
def switch(prediction):
  if (prediction==0):
    return "The person is not suffering from heart failure"
  else:
    return "The person is suffering from heart failure"
print(switch(prediction))
```
